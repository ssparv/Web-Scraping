# Web Scraping on Multiple Pages - Book Collection

```python
from bs4 import BeautifulSoup
import requests, openpyxl
import pandas as pd
```

```python
Collection =[]
for i in range(1,51):        # Total pages: 50
    source = requests.get(f'https://books.toscrape.com/catalogue/page-{i}.html') #f string used infront of url to make 'i' variable
    source.raise_for_status()

    soup = BeautifulSoup(source.text,'html.parser')

    ol = soup.find('ol')
    books = ol.find_all('article',class_='product_pod')

    for book in books:
        img = book.find('img')
        name = img.attrs['alt']
        Starrate = book.find('p')
        rate = Starrate.attrs['class'][1]
        price = book.find('div',class_='product_price').p.text
        price = float(price[2:])
        Availability = book.find('div',class_='product_price')
        Availability = Availability.find('p',class_='instock availability').get_text(strip=True)
        print(name, rate, price, Availability)
        Collection.append([name, rate, price, Availability])


```

```python
df = pd.DataFrame(Collection,columns = ['Name', 'Star_Rating', 'Price', 'Availability'])
df.to_csv('Book Collection.csv')
```
